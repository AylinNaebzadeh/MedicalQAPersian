```python
import os
import json
import random
from collections import defaultdict
from glob import glob
```

```python
def load_questions(data_directory):
    """Loads all questions.json files from the given directory."""
    all_questions = []
    
    json_files = glob(os.path.join(data_directory, "*", "questions.json"))

    print(f"Found {len(json_files)} JSON files.")  

    for file_path in json_files:
        with open(file_path, "r", encoding="utf-8") as f:
            questions = json.load(f)
            print(f"Loaded {len(questions)} questions from {file_path}")  
            all_questions.extend(questions)
    
    print(f"Total collected questions: {len(all_questions)}")  
    return all_questions
```

```python
def split_data(data, train_ratio=0.7, dev_ratio=0.15, test_ratio=0.15):
    """Splits data into train, dev, and test sets."""
    random.shuffle(data)  # Shuffle data before splitting
    total = len(data)

    train_size = int(total * train_ratio)
    dev_size = int(total * dev_ratio)
    test_size = total - (train_size + dev_size)  

    print(f"Splitting: {train_size} train, {dev_size} dev, {test_size} test")  

    train_data = data[:train_size]
    dev_data = data[train_size:train_size + dev_size]
    test_data = data[train_size + dev_size:]

    return train_data, dev_data, test_data
```

```python
def save_json(data, file_path):
    """Saves data to a JSON file."""
    with open(file_path, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

# BASIC MEDICAL SCIENCES SPLIT

```python
def main(data_directory, output_directory):
    questions = load_questions(data_directory)
    
    print(f"Total number of questions: {len(questions)}")

    if len(questions) == 0:
        print("No questions found. Exiting.")
        return

    train_data, dev_data, test_data = split_data(questions)

    train_file = os.path.join(output_directory, "basic_medical_sciences_train.json")
    dev_file = os.path.join(output_directory, "basic_medical_sciences_dev.json")
    test_file = os.path.join(output_directory, "basic_medical_sciences_test.json")

    save_json(train_data, train_file)
    save_json(dev_data, dev_file)
    save_json(test_data, test_file)

    print(f"BASIC MEDICAL SCIENCES: Train dataset saved: {len(train_data)} questions")
    print(f"BASIC MEDICAL SCIENCES: Dev dataset saved: {len(dev_data)} questions")
    print(f"BASIC MEDICAL SCIENCES: Test dataset saved: {len(test_data)} questions")
```

```python
data_directory = "/kaggle/input/medicalqapersian/data/data/basic_medical_sciences/"
output_directory = "/kaggle/working/"
main(data_directory, output_directory)
```
